```python
import cv2
import matplotlib.pyplot as plt

from skimage.feature import hog
from common import *
%matplotlib inline
```

```python
car_image = read_image('data/vehicles/GTI_Left/image0253.png')
plt.imshow(car_image);
```

```python
car_luv = convert_colorspace(car_image, 'LUV')

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(car_luv[:, :, 0], 'gray')
axes[0].set_title('[L] channel')

axes[1].imshow(car_luv[:, :, 1], 'gray')
axes[1].set_title('[U] channel')

axes[2].imshow(car_luv[:, :, 2], 'gray')
axes[2].set_title('[V] channel')
fig.tight_layout()
fig.savefig('./output_images/luv_channels.png')
```

```python
hog1, hog1_fig  = get_hog_features(car_luv[:, :, 0], orient=9, pix_per_cell=8, cell_per_block=2, vis=True)
hog2, hog2_fig  = get_hog_features(car_luv[:, :, 1], orient=9, pix_per_cell=8, cell_per_block=2, vis=True)
hog3, hog3_fig  = get_hog_features(car_luv[:, :, 2], orient=9, pix_per_cell=8, cell_per_block=2, vis=True)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(hog1_fig, 'gray')
axes[0].set_title('[L] channel')

axes[1].imshow(hog2_fig, 'gray')
axes[1].set_title('[U] channel')

axes[2].imshow(hog3_fig, 'gray')
axes[2].set_title('[V] channel')
fig.tight_layout()
fig.savefig('./output_images/hog_feature_extracted.png')
```
